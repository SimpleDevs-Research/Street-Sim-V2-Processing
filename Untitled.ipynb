```python
import os
import numpy as np
import pandas as pd
import pylab

from helpers import get_immediate_files
```

```python
_SAMPLES_DIR = 'data/'
_PDIR = [os.path.join(_SAMPLES_DIR, p) for p in ['P1','P2','P3','P4','P5','P6']]
_PFILES = {p : get_immediate_files(p) for p in _PDIR}
```

```python
def raycast_processing(raw_value):
    if len(raw_value) == 0: return '','','',''
    divided = raw_value.split("-")
    values = divided[0].split('.')
    values.append(divided[1])
    return values
```

# P1

```python
df = pd.read_csv('data/P1/eye_targets.csv')
df['raycast_hit_name'] = df['raycast_hit_name'].convert_dtypes()
df['raycast_hit_name'] = df['raycast_hit_name'].fillna('')
df
```

```python
grouped = df.groupby('raycast_hit_name', sort=False).first()[['unix_ms','rel_timestamp','frame']]
gazes = grouped.rename(columns={
    'unix_ms': 'start_unix_ms',
    'rel_timestamp': 'start_rel_timestamp',
    'frame': 'start_frame'
}).reset_index()

gazes['end_unix_ms'] = gazes['start_unix_ms'].shift(-1)  # Next group's start
gazes['end_rel_timestamp'] = gazes['start_rel_timestamp'].shift(-1)  # Next group's start
gazes['end_frame'] = gazes['start_frame'].shift(-1) # Next group's start


last_row = df.iloc[-1]
gazes.iloc[-1, gazes.columns.get_loc('end_unix_ms')] = last_row['unix_ms']
gazes.iloc[-1, gazes.columns.get_loc('end_rel_timestamp')] = last_row['rel_timestamp']
gazes.iloc[-1, gazes.columns.get_loc('end_frame')] = last_row['frame']

gazes['duration_frame'] = gazes['end_frame'] - gazes['start_frame']
gazes['duration_unix_ms'] = gazes['end_unix_ms'] - gazes['start_unix_ms']
gazes['duration_rel_timestamp'] = gazes['end_rel_timestamp'] - gazes['start_rel_timestamp']

gazes[['raycast_target_category','raycast_target_type','raycast_target_subtype','raycast_target_name']] = gazes.apply(
    lambda r: raycast_processing(r['raycast_hit_name']), 
    axis='columns', 
    result_type='expand'
)

display(gazes)
```

```python
# Get unique values in `raycast_target_category`
target_categories = sorted(gazes['raycast_target_category'].unique())

# Draw boxplot
bp = gazes.boxplot(column='duration_rel_timestamp', by='raycast_target_category', grid=False, color='blue')

# Add data on top
for i in range(0,len(target_categories)):
    c = target_categories[i]
    y = gazes['duration_rel_timestamp'][gazes['raycast_target_category'] == c].dropna()
    # Add some random "jitter" to the x-axis
    x = np.random.normal(i+1, 0.04, size=len(y))
    pylab.plot(x, y, 'r.', alpha=0.5)
```

```python
# Only look at vehicles.
vehicles = gazes[gazes['raycast_target_category']=='Vehicle']
target_subtypes = sorted(vehicles['raycast_target_subtype'].unique())

# Draw boxplot
bp = vehicles.boxplot(column='duration_rel_timestamp', by='raycast_target_subtype', grid=False, color='blue', rot=90)

# Add data on top
for i in range(0,len(target_subtypes)):
    c = target_subtypes[i]
    y = vehicles['duration_rel_timestamp'][vehicles['raycast_target_subtype'] == c].dropna()
    # Add some random "jitter" to the x-axis
    x = np.random.normal(i+1, 0.04, size=len(y))
    pylab.plot(x, y, 'r.', alpha=0.5)
```
